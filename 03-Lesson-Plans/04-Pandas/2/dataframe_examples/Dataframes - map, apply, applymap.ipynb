```python
import pandas as pd
import numpy as np
import random
```

```python
def create_random_progress_bar():
    """Returns a random progress bar with 5 to 40 * representing progress"""
    random_number = random.randint(5,40)
    random_list_of_starts = ["*" for number in range(random_number)]
    random_progress_bar = "".join(random_list_of_starts)
    return random_progress_bar
```

### Create a column with hours spent on challenge as a number

Add a column with a number that reflects how may hours a student spent on that challenge

### Method 1) Iterate over dataframe and create new list with length for each record

There are multiple ways to iterate over a dataframe:
- Use range with length of dataframe, use iloc, or loc to access value
- Use df.iterrows() - it returns index, and row for each record
- Use df.itertuples() - it returns tuple with content for each record




```python
# create a function that returns length of a string
def length_of_str(text):
    """Given text, return its length"""
    length = len(text)
    return length

# test function
length_of_str("*********")
```

Loop through each record in dataframe and build a list with length of hours_spent_on_challenge

```python
hours_on_challenge_list = []
for index in range(len(df)):    
    # Get row of data for this index
    row = df.iloc[index]
    # assign hours_spent_on_challenge to variable
    hours_extracted = row['hours_spent_on_challenge']     
    hours_extracted_as_number = length_of_str(hours_extracted)
    hours_on_challenge_list.append(hours_extracted_as_number)
```

```python
# Create a copy of df before assigning
df_assign_list = df.copy()
df_assign_list['hours_spent'] = hours_on_challenge_list
df_assign_list.head(10)
```

### Method 2) Use map method

map function only available with Series
- pass a dictionary to replace one value for another {1:'One', 2: 'Two', 3: 'Three'} 
- pass a function to apply to each element in the Series


```python
# Change challenge_number to alpha
# Create a copy of df before assigning
df_assign_map = df.copy()
df_assign_map['challenge_number'] = df_assign_map['challenge_number'].map({1:'One', 2: 'Two', 3: 'Three'})
df_assign_map.head(10)
```

![Series](images/assign_map.jpg)

```python
# map len function on each element
df_assign_map['hours_spent'] = df_assign_map['hours_spent_on_challenge'].map(len)
df_assign_map.head(10)
```

### Method 3) Use apply method

apply function available with Series and Pandas
- when apply is used with Series no need to set axis
- when apply is used with DataFrame it applies function to either axis on df
 - axis = 0 rows direction
 - axis = 1 columns direction


```python
# apply to Series
df_assign_apply = df.copy()
df_assign_apply['hours_spent'] = df_assign_map['hours_spent_on_challenge'].apply(length_of_str)
df_assign_apply.head(10)
```

![apply](images/assign_apply_function.jpg)

```python
# apply to DataFrame using an anonymous function
df_assign_apply_df = df.copy()
df_assign_apply_df['hours_spent'] = df_assign_apply_df.apply(lambda row: length_of_str(row['hours_spent_on_challenge']), axis=1)
df_assign_apply_df.head()
```

### Method 4) Use applymap

apply function to every element of a dataframe

```python
# apply to DataFrame using an applymap
# Add dashes beginning and end of each element
df_assign_applymap = df.copy()
df_assign_applymap.applymap("--{}--".format).head()
```

```python
df_assign_applymap['hours_spent'] = df_assign_applymap.loc[:,"hours_spent_on_challenge":].applymap(length_of_str)
```
