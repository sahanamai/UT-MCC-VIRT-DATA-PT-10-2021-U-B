```python
# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Random data

# X-Axis
x_data = np.linspace(0, 100, 25)
y_data = np.random.rand(25)
y_data
```

## State-based (MATLAB method)

Simple cases of programmatic plot generation. use plt.XXX

```python
plt.figure(facecolor='lightgrey')
# Draw on the first axes of a two by two (rows, columns) figure
plt.subplot(2,2,1)
plt.plot(x_data, y_data, 'r-')
# Draw on the second axes of a two by two (rows, columns) figure
plt.subplot(2,2,2)
plt.plot(x_data, y_data, 'b-')
# Draw on the fourth axes of a two by two (rows, columns) figure
plt.subplot(2,2,4)
plt.plot(x_data, y_data, 'g-')
plt.show()
```

## Object Oriented API is recommended for more complex plots


```python
fig, ax = plt.subplots(2,2)
fig.set_facecolor('lightgrey')

ax[0,0].plot(x_data, y_data, 'r-')
ax[0,1].plot(x_data, y_data, 'b-')
# Delete axes at 1,0 (row, column)
fig.delaxes(ax[1,0])
ax[1,1].plot(x_data, y_data, 'g-')
```
